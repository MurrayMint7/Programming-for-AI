```python
import plotly.express as px
import pandas as pd
```

```python
#Read in the data file

fifa = pd.read_csv('FifaGeo.csv')
```

```python
#Take a look at the data

fifa.head()
```

```python
#Create the scatter showing Potential matched against Age
px.scatter(fifa, 'Age', 'Potential', hover_data = ['Name','Club', 'Nationality'])
```

The plot does not look like the plot on the Activity sheet. What is the difference? What can be done to make it like the one on the Activity sheet?

```python
t1fifa = fifa.groupby('Age').mean(numeric_only=True).reset_index()
t1fifa.head()
```

```python
#Create the scatter showing Potential matched against Age
px.scatter(t1fifa, 'Age', 'Potential', hover_data = "Wage")
```

Note - it does not make sense to have the  hover data about individual attributes when presenting the graph using average values per age. The hover data in the above diagram was changed to "Wage"

```python
#Create the scatter showing Value matched against Age
px.scatter(fifa, 'Age', 'Value', hover_data = ['Name','Club', 'Nationality'])
```

```python
#Create the Geomap
px.scatter_geo(fifa, fifa['Lat'], fifa['Long'], hover_data = ['Club'])
```

```python
#Create a line plot for OverallRating of players
px.line(fifa,'Name', 'OverallRating')
```

The graph produced isn't quite like the plot on the activity sheet because OverallRating is not sorted.  The code below makes a graph more like the one on the activity sheet.

```python
df = fifa.sort_values(by="OverallRating", ascending=False)
fig = px.line(df, x="Name", y="OverallRating")
fig.show()
```

Next you will create the table.  For this you have call the Table method from plotly.graph_objects library directly to create a Table in a form plotly can understand.

```python
import plotly.graph_objects as go
```

```python
table_trace = go.Table (
    header=dict(values=['Name', 'Nationality', 'OverallRating']),
    cells=dict(values=[fifa['Name'], fifa['Nationality'], fifa['OverallRating']])
)
```

```python
tablefig = go.Figure(data=[table_trace])
```

```python
tablefig.show()
```

```python
#Create the Pie Chart
px.pie(fifa,'Nationality', 'Count')
```

The above graphic has not rendered in aesthetically pleasing way. The form can be improved by making it into a figure with the following commands.

```python
piefig = px.pie(fifa,'Nationality', 'Count')
```

```python
piefig.update_traces(textinfo='none')
```
