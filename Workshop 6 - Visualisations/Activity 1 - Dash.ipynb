This program uses Dash with Plotly to create a Dashboard. It is a very popular method for creating dashboards.


```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html, Dash

fifa = pd.read_csv('FifaGeo.csv')
```

```python
app = Dash()

table_trace = go.Table (
    header=dict(values=['Name', 'Nationality', 'OverallRating']),
    cells=dict(values=[fifa['Name'], fifa['Nationality'], fifa['OverallRating']])
)
tablefig = go.Figure(data=[table_trace])

piefig = px.pie(fifa,'Nationality', 'Count')
piefig.update_traces(textinfo='none')

app.layout = html.Div( [
             html.Div([dcc.Graph(figure=px.scatter(fifa, 'Age', 'Potential', hover_data= ['Name','Club', 'Nationality']))],
              style={'width': '33%', 'display': 'inline-block'}),
             html.Div([dcc.Graph(figure=px.scatter(fifa, 'Age', 'Value', hover_data= ['Name','Club', 'Nationality']))],
              style={'width': '33%', 'display': 'inline-block'}),
             html.Div([dcc.Graph(figure=px.scatter_geo(fifa, fifa['Lat'], fifa['Long'], hover_data= ['Club']))],
              style={'width': '33%', 'display': 'inline-block'}),
             html.Div([dcc.Graph(figure=px.bar(fifa,'Name', 'OverallRating'))],
              style={'width': '33%', 'display': 'inline-block'}),
             html.Div([dcc.Graph(figure=tablefig)],
              style={'width': '33%', 'display': 'inline-block'}),
             html.Div([dcc.Graph(figure=piefig)],
              style={'width': '33%', 'display': 'inline-block'})

               ])

if __name__ == '__main__':
    app.run(debug=True)
```
