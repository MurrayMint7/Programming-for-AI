```python
import numpy as np

import matplotlib.pyplot as plt

from numpy.polynomial.polynomial import Polynomial

# Generate some example data (change this as needed)

np.random.seed(42)

x = np.linspace(0, 10, 20)

y = -0.5 * x ** 3 + 4 * x ** 2 - 2 * x + 5 + 10 * np.exp(0.3 * x) + np.random.normal(0, 10, size=len(x))

# Set polynomial order (change this to 1, 2, 3, ..., 10 to see different fits)

poly_order = 2  # Change this value to adjust polynomial degree ### CHANGE ME

 

#Fit polynomial

coeffs = np.polyfit(x, y, poly_order)

poly = np.poly1d(coeffs)

#Generate fitted line

x_fit = np.linspace(min(x), max(x), 100)

y_fit = poly(x_fit)

 

#Plot data and polynomial fit

plt.scatter(x, y, label="Data", color="blue")

plt.plot(x_fit, y_fit, label=f"Polynomial Order {poly_order}", color="red", linewidth=2)

plt.xlabel("X")

plt.ylabel("Y")

plt.legend()

plt.title(f"Polynomial Regression (Order {poly_order})")

plt.show()

 

print(f"The equation of this Order {poly_order} polynomial is: ")

print('')

print(f"y=")

print(f"{np.poly1d(coeffs)}")

print('')

print("Note that the floating numbers are the power values that go after x. E.g. floating 2 means x^2")
```

```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error

 

#Generate data

np.random.seed(42)

x = np.linspace(0, 10, 20)

y = -0.5 * x ** 3 + 4 * x ** 2 - 2 * x + 5 + 10 * np.exp(0.3 * x) + np.random.normal(0, 10, size=len(x))

 

# Split data into training (70%) and test (30%)

train_size = int(0.7 * len(x))

x_train, x_test = x[:train_size], x[train_size:]

y_train, y_test = y[:train_size], y[train_size:]

 

#Store RMSE values

train_rmse_list = []

test_rmse_list = []

orders = list(range(0, 7))  # Polynomial degrees from 1 to X ### CHANGE ME

 

#Loop through polynomial orders

for poly_order in orders:

    coeffs = np.polyfit(x_train, y_train, poly_order)  # Fit polynomial

    poly = np.poly1d(coeffs)  # Create polynomial model

   

    # Predictions

    y_train_pred = poly(x_train)

    y_test_pred = poly(x_test)

 

    # Compute RMSE

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

 

    train_rmse_list.append(rmse_train)

    test_rmse_list.append(rmse_test)

 

# Plot RMSE vs. Polynomial Order

plt.figure(figsize=(8, 5))

plt.plot(orders, train_rmse_list, marker='o', label="Train RMSE", color='blue')

plt.plot(orders, test_rmse_list, marker='s', label="Test RMSE", color='red')

plt.xlabel("Polynomial Order")

plt.ylabel("RMSE")

plt.xticks(orders)

plt.legend()

plt.title("Bias-Variance Tradeoff: RMSE vs Polynomial Order")

plt.grid()

plt.show()
```
