```python
import matplotlib.pyplot as plt
import numpy as np

countries= ('Belgium', 'France', 'Germany', 'Spain', 'UK')
population=(11.48,66.99,83.02,46.94,66.65)
xpos = np.arange(len(countries))
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.bar(xpos,population, align='center', alpha=0.5, color=colors)
plt.xticks(xpos, countries)
plt.ylabel('Population')
plt.xlabel('Countries')
plt.title("Populations across countries")
plt.show()

Test1=np.array([59,52,44,51,42,42,41,45,27,63,54,44,50,47])
Test2=np.array([60,63,55,50,66,48,58,45,13,50,60,56,64,50])
colours = ['blue' if t1 < 50 else 'red' for t1 in Test1]
plt.scatter(Test1, Test2, alpha=0.5, color=colours)
plt.xlabel('Test1')
plt.ylabel('Test2')

degree = 4
coefficients = np.polyfit(Test1, Test2, degree)
polynomial = np.poly1d(coefficients)

# Generate x values for the regression line
x_values = np.linspace(min(Test1), max(Test1), 100)
y_values = polynomial(x_values)

plt.plot(x_values, y_values, color='red')

# Calculate predicted values and MSE
predicted_Test2 = polynomial(Test1)
mse = np.mean((Test2 - predicted_Test2) ** 2)

plt.text(30, 65, f'MSE: {mse:.2f}', fontsize=12, color='black')

plt.show()
```
