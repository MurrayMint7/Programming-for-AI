```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#set path
path = "./FIFAPLAYERS.csv"
#use pandas to read the CSV file
df = pd.read_csv(path, encoding="ISO-8859-1")

# Create histograms for age, value, and wage
columns_to_plot = ['Age', 'Value', 'Wage']

for column in columns_to_plot:
    if column in df.columns:
        plt.figure(figsize=(8, 6))
        
        if column == 'Value' or column == 'Wage':         
            # Format x-axis to show values in millions or thousands
            ax = plt.gca()
            if column == 'Value':
                plt.hist(df[column].dropna(), bins=100, color='blue', alpha=0.7, edgecolor='black', range=(0, 50000000))
                ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))
                plt.xlabel(f'{column} (in Millions €)')
            else:  # Wage
                plt.hist(df[column].dropna(), bins=50, color='blue', alpha=0.7, edgecolor='black', range=(0, 300000))
                ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x/1e3:.0f}K'))
                plt.xlabel(f'{column} (in Thousands €)')
        else:
            plt.hist(df[column].dropna(), bins=30, color='blue', alpha=0.7, edgecolor='black')
            plt.xlabel(column)
        
        plt.title(f'Distribution of {column}')
        plt.ylabel('Frequency')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
    else:
        print(f"Column '{column}' not found in the dataset.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#set path
path = "./FIFAPLAYERS.csv"
#use pandas to read the CSV file
df = pd.read_csv(path, encoding="ISO-8859-1")

# Normalize data for better visualization
df['Value'] = df['Value'] / 1e6  # Convert to millions
df['Wage'] = df['Wage'] / 1e3  # Convert to thousands

# Remove rows with NaN values in 'Value' or 'Wage'
df = df.dropna(subset=['Value', 'Wage'])

plt.scatter(df['Value'], df['Wage'], alpha=0.5, color='blue')
plt.xlabel('Value (in Millions €)')
plt.ylabel('Wage (in Thousands €)')

m, b = np.polyfit(df['Value'], df['Wage'], 1)
print(f"Slope: {m}, Intercept: {b}")

prediction = m * df['Value'] + b
x_values = np.array([df['Value'].min(), df['Value'].max()])
y_values = m * x_values + b

error = np.mean((prediction - df['Wage'])**2)
plt.text(df['Value'].min(), df['Wage'].max(), f'MSE: {error:.2f}', fontsize=12, color='black')

plt.plot(x_values, y_values, color='red')

plt.show()
```
